```python
def Preprocessing(df):
    df['Race'].fillna(df['Race'].mode()[0],inplace=True)
    df['DiabetesMellitus'].fillna("NO DM",inplace=True)
    df['ChronicKidneyDisease'].fillna("NO CKD",inplace=True)
    df['Anemia'].fillna("NO Anemia",inplace=True)
    df['Depression '].fillna("NO Depression",inplace=True)
    df['ChronicObstructivePulmonaryDisease'].fillna("NO COPD",inplace=True)
    df['PatientId']   = pd.to_numeric(df['PatientId'])
    df['EncounterId'] = pd.to_numeric(df['EncounterId'],errors='coerce')
    df.isnull().sum()
    df['EncounterId'].fillna(0,inplace=True)
    le = LabelEncoder()
    categ = ['Race','DischargeDisposision','Gender','DiabetesMellitus','ChronicKidneyDisease','Anemia','Depression ','ChronicObstructivePulmonaryDisease','ReadmissionWithin_90Days']
    df[categ] = df[categ].apply(le.fit_transform)

    return df
```

```python
df = Preprocessing(df)
df
```

```python
X = df.drop('ReadmissionWithin_90Days',axis=1)
y = df['ReadmissionWithin_90Days']
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2 , random_state = 60)
```

```python
rm = RandomForestClassifier(n_estimators = 10, max_depth=25, criterion = "gini", min_samples_split=10)
rm.fit(X_train, y_train)
pred = rm.predict(X_test)
print ("Accuracy : " , accuracy_score(y_test,pred)*100)
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
```

GENETIC ALGORITHM IMPLEMENTATION

```python
# initialize population
def initialize_population(size):
    Arr = [True,False]
    population = np.random.choice(Arr,size=(size,56))
    return population
```

```python
def fitness_function(X_train,y_train):
    rm = RandomForestClassifier(n_estimators = 20, max_depth=25, criterion = "entropy", min_samples_split=10)
    rm.fit(X_train, y_train)
    return rm
```

Sorting accuracies W.R.T the boolean array

```python
def sorting_chromosome(S_features,accuracy):
    indexes = []
    so = sorted(accuracy,reverse=True)
    for i in range(len(accuracy)):
        for j in range(len(accuracy)):
            if so[i] == accuracy[j]:
                indexes.append(j)
    x = 0
    for i in range(len(accuracy)):
        S_features[i],S_features[indexes[i]] = S_features[indexes[i]],S_features[i]
        x+=1
    return S_features
```

Single point crossover

```python
def Crossover(S_features,r):
    split = np.random.randint(10,56)
    # split = 56//2
    r = r//2
    if (r % 2 == 1):
        r+=1
    S_features = S_features[:r]   # splitting half the features which gave lower accuraices
    for i in range(0,r,2):
        if i+1 == (r-1):
            break
        # crossover
        c1 = S_features[i,:split]
        c2 = S_features[i,split:]
        c3 = S_features[i+1,:split]
        c4 = S_features[i+1,split:]
        #joining the cross values
        child1 = np.concatenate((c1,c4),axis=0)
        child2 = np.concatenate((c2,c3),axis=0)
        child1 = child1.reshape(56,1)
        child2 = child2.reshape(56,1)
        S1 = np.concatenate((child1,child2),axis=0)
        S1 = S1.reshape(2,56)
        l = np.concatenate((S_features,S1),axis=0)

    l = l[:r]
        #Mutation
    n = np.random.uniform(0.0,1.0)
    if n > 0.5:
        l = Mutation(l)
    l = np.concatenate((S_features,l),axis=0)

    return l
```

Mutation

```python
def Mutation(l):
    for r in range(len(l)):
        random = np.random.randint(56,size=10)
        # print(l[0])
        # print(random)
        for i in random:
            if l[r][i] == False:
                l[r][i] = True
            else:
                l[r][i] = False
    return l
```

Genetic Algorithm Implementation

```python
r = 30
S_features = initialize_population(r) # initializing population
for l in range(30):
    accuracy = []
    for i in range(0,r):
        sel = X.loc[:,S_features[i]]
        X_train, X_test, y_train, y_test = train_test_split(sel, y , test_size = 0.2 , random_state = 60)
        rm = fitness_function(X_train,y_train)
        pred=rm.predict(X_test)
        a = accuracy_score(y_test,pred)*100
        accuracy.append(a)

    print('max : ',max(accuracy))


    # Sorting accuracies W.R.T the boolean array
    S_features = sorting_chromosome(S_features,accuracy)

    # Single point crossover
    l = Crossover(S_features,r)

    S_features = l

    # termination condition
    if max(accuracy) > 73:
        print ("Accuracy : " , accuracy_score(y_test,pred)*100)
        print("Report : \n", classification_report(y_test, pred))
        print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
        break

```
